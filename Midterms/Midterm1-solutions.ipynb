We can define a spiral parametrically by

$$\begin{align}
    x_1(t) &= t^{1/2} \cos(t)  \\
    y_1(t) &= t^{1/2} \sin(t)
\end{align}$$

We can also define a cardioid via

$$\begin{align}
    x_2(t) &= 2a[1-\cos(t)]\cos(t)  \\
    y_2(t) &= 2a[1-\cos(t)]\sin(t) 
\end{align}$$

By letting $a=1.2$, plot both of these curves on the same plot with $t=0\dots4\pi$. Add a grid to the plot and a lengend showing 'spiral' and 'cardiod'

```python
a=1.2

t = np.linspace(0,4*np.pi,1000)
x1 = t**1/2*np.cos(t)
y1 = t**1/2*np.sin(t)

x2 = 2*a*(1-np.cos(t))*np.cos(t)
y2 = 2*a*(1-np.cos(t))*np.sin(t)

plt.grid(True)
plt.plot(x1,y1);
plt.plot(x2,y2);
plt.legend(['spiral','cardioid']);
```

## Q3a [4 marks]

Using the below `SecantMethod()` function, numerically find all the roots of $f(x) = 6\sin(x)-x^2 + 1$ to within an error $<10^{-6}$. 

Hints:

- It may help to plot the function to find good initial guesses.
- Check the answers you find give $f(x_*)$ very near zero

```python
def SecantMethod(f, x0, x1, tol):
    i = 0
    xim1 = x0
    xi   = x1
    fi = 1
    while np.abs(fi) > tol:
        fi   = f(xi)
        xi1  = xi - fi*(xi - xim1)/(fi - f(xim1))
        xim1 = xi
        xi   = xi1
        
        i += 1

    return xi1
```

```python
def f(x):
    return 6*np.sin(x) - x**2 + 1
```

```python
x = np.linspace(-10,10,100)
y = f(x)
plt.grid(True)
plt.plot(x,y);
```

```python
x0 = SecantMethod(f,-0.1,0.0,1e-6)
x1 = SecantMethod(f, 2.5,2.6,1e-6)
print(x0, x1)
print(f(x0),f(x1))
```

## Q3b [5 marks]

Use the below FindRootBisection method in this question -- note the function only works when $f(a) < 0 < f(b)$

```python
def FindRootBisection(f, a, b, tol):
    left = a
    right = b
    while right-left > tol:
        midpoint = (left+right)/2
        if(f(midpoint) < 0):
            left = midpoint
        else:
            right = midpoint
    return midpoint
```

# Question 4: interpolation [5 marks]

In this question you should use the ``LagrangePoly`` function below

```python
# For the arguements: 
#     x is the data point (or array of points) to evaluate the intepolaring polynomial at
#     data is the data to be interpolated
def LagrangePoly(x, data):
    n = data.shape[0] - 1
    i = 1
    fn = 0
    while i <= n + 1:
        j = 1
        Li = 1
        while j <= n+1:
            if(j == i):
                j += 1
                continue
            Li *= (x - data[j-1,0])/(data[i-1,0] - data[j-1,0])
            j += 1
        fn += data[i-1,1]*Li
        i += 1
    return fn
```
